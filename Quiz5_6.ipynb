```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt


data = pd.read_csv('pesticides.csv')

#   
mexico_data = data[data['Area'] == 'Mexico']

X = mexico_data[['Year']]
y = mexico_data['Value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#  
print("Model efficiency for Mexico:", r2_score(y_test, y_pred))


plt.scatter(X_test, y_test, color='black', label='Actual Data')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Regression Line')
plt.title('SLR Model for pesticides use in Mexico')
plt.xlabel('Years')
plt.ylabel('Amount of pesticides used in Mexico (tonnes)')

#   
for i, year in enumerate(X_test['Year']):
    plt.annotate(str(year), (X_test.iloc[i, 0], y_test.iloc[i]), textcoords="offset points", xytext=(0,5), ha='center')

plt.legend()
plt.show()
```
